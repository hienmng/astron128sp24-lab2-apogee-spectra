## Problem 1 (Downloading and Plotting APOGEE Spectra)

Multiple Visit spectra are combined spectra of stars observed multipled times to increase S/N ratio. The Barycentric frame is the frame of the center of the mass of the solar system. Barycentric correction would refer correction in which Earth's radial velocity relative to the observed source is subtracted from the observed radial velocity. The Doppler shift would be different for each visit because Earth's orbit will constantly change the observed radial velocity of the star. The star could also be in a dynamic system such as a binary which will influence the observed radial velocity.

```python
def plot_spectra(file_name):
    f = fits.open(file_name)
    spectrum = f[1].data[0]
    error = f[2].data[0]
    mask = f[3].data[0]

    CRVAL = f[1].header['CRVAL1']
    CRDELT1 = f[1].header['CDELT1'] 
    wavelength = 10**np.arange(CRVAL, CRVAL + len(spectrum)*CRDELT1, CRDELT1)

    plt.figure(figsize = (14,7))
    plt.grid()
    plt.plot(wavelength, spectrum)
    plt.title(f'Wavelength vs. Flux, {file_name}', size = 20)
    plt.xlabel('Wavelength (Angstroms)', size = 20)
    plt.ylabel('Flux ($10^{-17}$ erg/s/cm^2/Ang)', size = 20)
    
    return wavelength, spectrum, error, mask

def get_spectra_data(file_name):
    f = fits.open(file_name)
    spectrum = f[1].data
    if(len(spectrum) == 8575):
        spectrum = f[1].data
        error = f[2].data
        mask = f[3].data
    else:
        spectrum = f[1].data[0]
        error = f[2].data[0]
        mask = f[3].data[0]
    
    CRVAL = f[1].header['CRVAL1']
    CRDELT1 = f[1].header['CDELT1'] 
    wavelength = 10**np.arange(CRVAL, CRVAL + len(spectrum)*CRDELT1, CRDELT1)
    
    
    return wavelength, spectrum, error, mask

#get_spectra_data(bad)
```

```python
def normalize_spectra(file_name):
    
    clean_spectra, clean_wavelength, clean_errors = get_clean_spectra(file_name)
    
    start1 = np.searchsorted(clean_wavelength, 15150)
    end1 = np.searchsorted(clean_wavelength, 15800)

    start2 = np.searchsorted(clean_wavelength, 15890)
    end2 = np.searchsorted(clean_wavelength, 16430)

    start3 = np.searchsorted(clean_wavelength, 16490)
    end3 = np.searchsorted(clean_wavelength, 16950)-1
    
      
    params1 = np.polyfit(clean_wavelength[start1:end1], clean_spectra[start1:end1], 3, w = clean_errors[start1:end1])

    params2 = np.polyfit(clean_wavelength[start2:end2], clean_spectra[start2:end2], 3, w = clean_errors[start2:end2])

    params3 = np.polyfit(clean_wavelength[start3:end3], clean_spectra[start3:end3], 3, w = clean_errors[start3:end3])
    
    
    pseudo_continuum1 = np.polyval(params1, clean_wavelength[start1:end1])
    pseudo_continuum2 = np.polyval(params2, clean_wavelength[start2:end2])
    pseudo_continuum3 = np.polyval(params3, clean_wavelength[start3:end3])
    
    n1 = clean_spectra[start1:end1] / pseudo_continuum1 
    n2 = clean_spectra[start2:end2] / pseudo_continuum2
    n3 = clean_spectra[start3:end3] / pseudo_continuum3
    
    e1 = clean_errors[start1:end1] / pseudo_continuum1
    e2 = clean_errors[start2:end2] / pseudo_continuum2
    e3 = clean_errors[start3:end3] / pseudo_continuum3
    
    normalized_spectrum = np.array(n1)
    normalized_spectrum = np.append(normalized_spectrum, n2)
    normalized_spectrum = np.append(normalized_spectrum, n3)
    
    normalized_errors = np.array(e1)
    normalized_errors = np.append(normalized_errors, e2)
    normalized_errors = np.append(normalized_errors, e3)
    
    new_wavelength = np.array(clean_wavelength[start1:end1])
    new_wavelength = np.append(new_wavelength, clean_wavelength[start2:end2])
    new_wavelength = np.append(new_wavelength, clean_wavelength[start3:end3])
    
    
    return normalized_spectrum, new_wavelength, normalized_errors

def normalize_spectra_2(file_name):
    wavelength, spectrum, errors, masks = get_spectra_data(file_name)
    clean_spectra, clean_wavelength, clean_errors = get_clean_spectra(file_name)
    
    params = np.polyfit(clean_wavelength, clean_spectra, 3, w = clean_errors)
    pseudo_continuum = np.polyval(params, wavelength)
    
    spectrum = spectrum / pseudo_continuum
    
    return spectrum, wavelength
    
    
    
def plot_normalized_spectra(file_name):
    
    #wavelength, spectrum, errors, masks = get_spectra_data(file_name)
    #clean_spectra, clean_wavelength, clean_errors = get_clean_spectra(file_name)
    
    normalized_clean_spectrum, new_wavelength, normalized_errors = normalize_spectra(file_name)
    
    
    plt.figure(figsize = (20,7))
    plt.grid()
    plt.plot(new_wavelength, normalized_clean_spectrum, label = 'normalized spectrum')
    plt.title(f'Normalized Spectrum, {file_name}', size = 20)
    plt.xlabel('Wavelength (Angstroms)', size = 20)
    plt.ylabel('Normalized Flux', size = 20)
    
```

```python
import random


#solve for coefficents

def get_coef(wavelength_bin, s):
    
    f_data = spectra_list[:, wavelength_bin]
    sigma = np.sqrt(errors_list[:, wavelength_bin]**2 + s**2)
    
    #weighted least squares optimization 
    
    W = np.sqrt(np.diag(1/sigma**2))
    X_w = np.dot(W,X)
    f_data_w = np.dot(f_data,W)
    thetas = np.linalg.lstsq(X_w, f_data_w, rcond = -1)[0]
            
        
    return thetas

#compute likelihood as a function of intrinsic scatter

def likelihood(f_data, f_model, sigma, s):
    
    #f_data = np.random.normal(size = 100)
    #f_model = np.zeros_like(f_data)
    #sigma = np.zeros_like(f_data)
    
    L = 0

    for i in range(len(f_data)):

        L += -0.5 * ((f_data[i]-f_model[i])**2)/(sigma[i]**2 + s**2) - 0.5*np.log((sigma[i]**2 + s**2))
    
    return L


def get_intrinsic_scatter(wavelength_bin):
    
    s_grid = 10**np.linspace(-5, 0,15)
    
    #thetas = get_coef(wavelength_bin)
    
    
    f_data = spectra_list[:, wavelength_bin]
    sigma = errors_list[:, wavelength_bin]
    
    #f_model = np.matmul(X, thetas)
    
    
    L_list = []
    
    for s in s_grid:
        thetas = get_coef(wavelength_bin, s)
        f_model = np.matmul(X, thetas)
        
        L = likelihood(f_data, f_model, sigma, s)
        L_list.append(L)
    
    index = np.argmax(L_list)
    
    return s_grid[index], thetas

    
```

```python
from scipy.optimize import curve_fit

def func(thetas, teff, logg, fe_h, mg_fe, si_fe):
    
    teff = teff - teff_mean
    logg = logg-logg_mean
    fe_h = fe_h-fe_h_mean
    mg_fe = mg_fe-mg_fe_mean
    si_fe = si_fe-si_fe_mean
    
    row = [1, teff, logg, fe_h, mg_fe, si_fe, 
       teff**2, teff*logg, teff*fe_h, teff*mg_fe, teff*si_fe,
      logg**2, logg*fe_h, logg*mg_fe, logg*si_fe,
       fe_h**2, fe_h*mg_fe, fe_h*si_fe,
       mg_fe**2, mg_fe*si_fe, si_fe**2]
    
    return np.matmul(thetas, row)


teff_bf_list = []
logg_bf_list = []
fe_h_bf_list = []
mg_fe_bf_list = []
si_fe_bf_list = []

teff_bf_err = []
logg_bf_err = []
fe_h_bf_err = []
mg_fe_bf_err = []
si_fe_bf_err = []

teff_test_list = []
logg_test_list = []
fe_h_test_list = []
mg_fe_test_list = []
si_fe_test_list = []

teff_test_err_list = []
logg_test_err_list = []
fe_h_test_err_list = []
mg_fe_test_err_list = []
si_fe_test_err_list = []




x = 0
for index in range(len(label_list_test)):
    
    sigma = np.sqrt(errors_list_test[index]**2 + s_intrinsic_list[index]**2)
    
    popt, pcov = curve_fit(func, thetas_matrix, spectra_list_test[index], 
              p0 = [teff_mean, logg_mean, fe_h_mean, mg_fe_mean, si_fe_mean], 
              sigma = sigma, method ='trf')
    
    teff_bf_list.append(popt[0])
    logg_bf_list.append(popt[1])
    fe_h_bf_list.append(popt[2])
    mg_fe_bf_list.append(popt[3])
    si_fe_bf_list.append(popt[4])
    
    #errors
    teff_bf_err.append(pcov[0,0])
    logg_bf_err.append(pcov[1,1])
    fe_h_bf_err.append(pcov[2,2])
    mg_fe_bf_err.append(pcov[3,3])
    si_fe_bf_err.append(pcov[4,4])
    
    teff_test_list.append(label_list_test.iloc[index]['TEFF'])
    logg_test_list.append(label_list_test.iloc[index]['LOGG'])
    fe_h_test_list.append(label_list_test.iloc[index]['FE_H'])
    mg_fe_test_list.append(label_list_test.iloc[index]['MG_FE'])
    si_fe_test_list.append(label_list_test.iloc[index]['SI_FE'])
    
    teff_test_err_list.append(label_list_test.iloc[index]['TEFF_ERR'])
    logg_test_err_list.append(label_list_test.iloc[index]['LOGG_ERR'])
    fe_h_test_err_list.append(label_list_test.iloc[index]['FE_H_ERR'])
    mg_fe_test_err_list.append(label_list_test.iloc[index]['MG_FE_ERR'])
    si_fe_test_err_list.append(label_list_test.iloc[index]['SI_FE_ERR'])
    
    print(x)
    x+=1




```
